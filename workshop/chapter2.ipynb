# step 2 select the Algorithm(logistic regression)

# step 3 bulid a model

```python
from sklearn.linear_model import LogisticRegression
```

```python
X = credit_data.loc[: , 'for_wor_A201' : 'dependents']
Y = credit_data['default']
```

```python
logModel = LogisticRegression(solver = 'liblinear')
logModel = logModel.fit(X,Y)
```

```python
logModel.score(X,Y)
```

# step 4 check model quality

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
logModel = logModel.fit(X_train, Y_train)
predictions = logModel.predict(X_test)
predictions
```

```python
logModel.score(X_test, Y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, predictions)
```

```python
print(logModel.intercept_[0])
print(*logModel.coef_[0], sep='\n')
```
