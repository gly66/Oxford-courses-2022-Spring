```python
import numpy as np
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
df = df = pd.read_csv("word_count.csv")
```

```python
df.fillna("空",inplace=True)
```

```python
df['word']
```

```python
text = df.word.to_list()
text = ' '.join(text).lower()
wordcloud = WordCloud(stopwords = STOPWORDS,
                      collocations=True, font_path = 'simhei.ttf', background_color='white', width = 2000,height=1000).generate(text)

plt.figure(figsize=(75,75))
plt.imshow(wordcloud,interpolation='bilInear')
plt.axis('off')
plt.show()
```
